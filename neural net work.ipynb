```python
#以下是两层神经网络  将一个一千维的向量  转换成10维的
#N是样本数量
#D_in 输入维度，可能是特徵值
#H 隐藏层维度
#D_out 输出数量， 由题目确定
import numpy as np

# N,D_in,H,D_out = 64 , 1000 , 100 ,10
N,D_in,H,D_out = 64 , 1000 , 100 ,10

#随机创建一些训练数据
x = torch.randn(N,D_in)
y = torch.randn(N,D_out)

w1 = torch.randn(D_in,H)
w2 = torch.randn(H,D_out)

learning_rate = 1e-6  #学习率
for t in range(500):   #迭代500次
    #forward pass
    h = x.mm(w1)  #N * H
    h_relu = h.clamp(min=0)  # N * H
    y_pred = h_relu.mm(w2)

    #compute loss
    loss = (y_pred - y).pow(2).sum().item()
    print(t,loss)
    #backward pass
    #comput gradient   手动算梯度
    grad_y_pred = 2.0 * (y_pred - y)
    grad_w2 = h_relu.t().mm(grad_y_pred)
    grad_h_relu = grad_y_pred.mm(w2.t())
    grad_h = grad_h_relu.clone()
    grad_h[h<0] = 0
    grad_w1 = x.t().mm(grad_h)
    
    #update weights of w1 ,w2
    w1 -= learning_rate *grad_w1
    w2 -= learning_rate *grad_w2
```

```python
#N是样本数量
#D_in 输入维度，可能是特徵值
#H 隐藏层维度
#D_out 输出数量， 由题目确定
import numpy as np

# N,D_in,H,D_out = 64 , 1000 , 100 ,10
N,D_in,H,D_out = 64 , 1000 , 100 ,10

#随机创建一些训练数据
x = torch.randn(N,D_in)
y = torch.randn(N,D_out)
#cuda 加速
#w2 = torch.randn(H,D_out).to("cuda:0")
w1 = torch.randn(D_in,H,requires_grad = True)
w2 = torch.randn(H,D_out,requires_grad = True)

learning_rate = 1e-6  #学习率
for t in range(500):   #迭代500次
    #forward pass
    y_pred = x.mm(w1).clamp(min=0).mm(w2)  #N * H

    #compute loss
    loss = (y_pred - y).pow(2).sum()
    print(t,loss.item())
    #backward pass
    loss.backward()
    #comput gradient   手动算梯度
    with torch.no_grad():  #改循环随时清空grad  不然会出现乱七八糟的错误
        #update weights of w1 ,w2
        w1 -= learning_rate *w1.grad
        w2 -= learning_rate *w2.grad
        w1.grad.zero_()
        w2.grad.zero_()#清空grad，不然会出bug
```
