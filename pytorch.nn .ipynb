```python
import torch.nn as nn
#使用nn库来构建网络。
#N是样本数量
#D_in 输入维度，可能是特徵值
#H 隐藏层维度
#D_out 输出数量， 由题目确定
import numpy as np
import torch
# N,D_in,H,D_out = 64 , 1000 , 100 ,10
N,D_in,H,D_out = 64 , 1000 , 100 ,10

#随机创建一些训练数据
x = torch.randn(N,D_in)
y = torch.randn(N,D_out)
#cuda 加速
#w2 = torch.randn(H,D_out).to("cuda:0")
w1 = torch.randn(D_in,H,requires_grad = True)
w2 = torch.randn(H,D_out,requires_grad = True)

learning_rate = 1e-6  #学习率
for t in range(500):   #迭代500次
    #forward pass
    y_pred = x.mm(w1).clamp(min=0).mm(w2)  #N * H

    #compute loss
    loss = (y_pred - y).pow(2).sum()
    print(t,loss.item())
    #backward pass
    loss.backward()
    #comput gradient   手动算梯度
    with torch.no_grad():  #改循环随时清空grad  不然会出现乱七八糟的错误
        #update weights of w1 ,w2
        w1 -= learning_rate *w1.grad
        w2 -= learning_rate *w2.grad
        w1.grad.zero_()
        w2.grad.zero_()#清空grad，不然会出bug
        
#该模型还存在很多问题，可以简化如下
```

```python
#pytorch 总体框架

import torch.nn as nn
#使用nn库来构建网络。
#N是样本数量
#D_in 输入维度，可能是特徵值
#H 隐藏层维度
#D_out 输出数量， 由题目确定
import numpy as np
import torch
# N,D_in,H,D_out = 64 , 1000 , 100 ,10
N,D_in,H,D_out = 64 , 1000 , 100 ,10

#随机创建一些训练数据
x = torch.randn(N,D_in)
y = torch.randn(N,D_out)
#定义模型
class TwoLayerNet(torch.nn.Module):
    def __init__(self,D_in,H,D_out):
        super(TwoLayerNet,self).__init__()
        #defin the model architecture
        self.linear1 = torch.nn.Linear(D_in,H,bias=False)
        self.linear2 = torch.nn.Linear(H,D_out,bias=False)
    def forward(self,x):
        y_pred = self.linear2(self.linear1(x).clamp(min=0))
        return y_pred
        
Model = TwoLayerNet(D_in,H,D_out)

#定义lossfunction
loss_fn = nn.MSELoss(reduction='sum')

learning_rate = 1e-4#
#定义优化函数
optimizer = torch.optim.Adam(Model.parameters(),lr=learning_rate)
#用几个不同的优化函数

for t in range(500):   
    #forward pass
    y_pred = Model(x)#因为上面已经定义好了模型，  这一条就可以帮我们做model.forward（）  前向传播就做好了
    #y_pred = x.mm(w1).clamp(min=0).mm(w2)  #N * H

    #compute loss
    loss = loss_fn(y_pred , y)#同下
    #loss = (y_pred - y).pow(2).sum()
    print(t,loss.item())
    
    optimizer.zero_grad()
    #backward pass
    loss.backward()
    #comput gradient

    #updata model parameters 
    optimizer.step()


    
```
